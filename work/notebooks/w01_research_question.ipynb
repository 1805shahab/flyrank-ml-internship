## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
"""
# Research Question

Decision:
Which pages should be reviewed first for content refresh?

Unit of Analysis:
Each row represents one content page.

Output:
A ranked list of pages ordered by their refresh priority.

Action:
An SEO or content team reviews the highest-ranked pages to determine whether they should be updated, expanded, rewritten, monitored, or left unchanged.

Cost of a Wrong Recommendation:
If pages are ranked incorrectly, valuable content may continue losing traffic while time is spent reviewing pages that do not require updates. This reduces the efficiency of the content review process.

Why Machine Learning Can Help:
Content performance depends on several interacting factors such as impressions, CTR, ranking position, content age, and freshness. Machine learning may identify useful patterns that are difficult to capture using simple hand-written rules while still allowing human review before any action is taken.
"""
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os
import subprocess
import sys
import pandas as pd

REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if not os.path.isdir(REPO_DIR):
    subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)

os.chdir(REPO_DIR)

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Number of pages:", len(df))
print("Average CTR:", round(df["ctr"].mean(), 4))
print("Median content age (days):", df["content_age_days"].median())

print("\nTrend Direction Distribution:")
print(df["trend_direction"].value_counts())

"""
## Quick Look at the Data

The starter dataset contains **30,000 content pages**, providing enough examples to study content refresh opportunities.

The **average CTR is 0.5107**, indicating that pages have varying levels of user engagement, making CTR a useful signal to investigate alongside other metrics.

The **median content age is 236 days**, suggesting that many pages have not been updated for several months and may be candidates for review.

The trend distribution also shows that **16,262 pages are labeled as 'down'**, which is substantially more than the other categories. This suggests that prioritizing declining pages for review is a meaningful problem to investigate.
"""
```
